```python
import cv2
import matplotlib.pyplot as plt
import os
import math

# Get the list of output images
output_dir = "runs/detect/predict"
output_images = [f for f in os.listdir(output_dir) if f.endswith(('.jpg', '.png'))]

# Set grid size dynamically based on the number of images
num_images = len(output_images)
cols = 4  # Number of columns
rows = math.ceil(num_images / cols)

# Display the images in a grid
fig, axes = plt.subplots(rows, cols, figsize=(20, 20))  # Larger figure size
axes = axes.flatten()

for i, img_name in enumerate(output_images):
    img = cv2.imread(f"{output_dir}/{img_name}")
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    axes[i].imshow(img)
    axes[i].axis('off')
    axes[i].set_title(img_name, fontsize=12)

# Hide unused subplots if there are fewer images than grid slots
for i in range(num_images, len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()

```
